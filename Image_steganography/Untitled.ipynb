```python

from PIL import Image

def encode_image(image_path, message, output_path):
    image = Image.open(image_path)
    width, height = image.size
    max_message_size = width * height * 3 // 8
    if len(message) > max_message_size:
        raise ValueError("Message too large for image")
    
    # Convert the message to binary
    message_bits = ''.join(format(ord(char), '08b') for char in message)
    
    # Create a copy of the image to encode the message
    encoded_image = image.copy()
    
    # Encode the message into the image
    x, y = 0, 0
    for bit in message_bits:
        pixel = list(encoded_image.getpixel((x, y)))
        if bit == '0':
            pixel[0] = (pixel[0] & ~1)
        else:
            pixel[0] = (pixel[0] | 1)
        encoded_image.putpixel((x, y), tuple(pixel))
        x += 1
        if x == width:
            x = 0
            y += 1
    
    # Save the encoded image
    encoded_image.save(output_path)

def decode_image(image_path):
    image = Image.open(image_path)
    width, height = image.size
    
    # Extract the encoded message from the image
    message_bits = ''
    x, y = 0, 0
    while len(message_bits) < 8:
        pixel = image.getpixel((x, y))
        message_bits += str(pixel[0] & 1)
        x += 1
        if x == width:
            x = 0
            y += 1
    
    # Convert the binary message to text
    message = ''.join(chr(int(message_bits[i*8:i*8+8], 2)) for i in range(len(message_bits)//8))
    
    return message

def decode_image(image_path):
    image = Image.open(image_path)
    width, height = image.size
    
    # Extract the encoded message from the image
    message_bits = ''
    x, y = 0, 0
    while True:
        pixel = image.getpixel((x, y))
        message_bits += str(pixel[0] & 1)
        x += 1
        if x == width:
            x = 0
            y += 1
        if len(message_bits) >= 8 and message_bits[-8:] == '00000000':
            break
    
    # Convert the binary message to text
    message = ''
    for i in range(0, len(message_bits), 8):
        byte = message_bits[i:i+8]
        if byte == '00000000':
            break
        message += chr(int(byte, 2))
    
    return message

# Example usage
image_path = 'uchiha.png'
message = 'Hello, World!'
output_path = 'encoded_uchiha.png'

encode_image(image_path, message, output_path)
print(f"Message encoded and saved to {output_path}")

decoded_message = decode_image(output_path)
print(f"Decoded message: {decoded_message}")
#This code encodes a message into an image and decodes it back. Make sure to replace the image_path, message, and output_path variables with your own values.
#Note: This is a simple example of image steganography and not secure for practical use. In a real-world scenario, you should use a more secure encryption algorithm and consider additional factors like image size and message length.
```
